```python
from numpy import *
class ComplexNumber :
    #Task 1
    def __init__(self, real, imaginary): 
        self.real = real
        self.imaginary = imaginary

    #Task 2    
    def __add__(self, other):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        if isinstance(other, ComplexNumber):
            p2, q2 = other.real, other.imaginary
    #Task 8
        elif isinstance(other, float) & isinstance(self,ComplexNumber):
            p2, q2 = other, 0
            p1, q1 = self.real, self.imaginary
        elif isinstance(self, float) & isinstance(other,ComplexNumber):
            p1, p2 = self, 0
            p1, q1 = other.real, other.imaginary
        else:
            raise TypeError('Wrong type!')
        return ComplexNumber(p1 + p2, q1+q2)
    def __radd__(self, other):
        return self + other 
    
    def __sub__(self, other):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        p2, q2 = other.real, other.imaginary
        return ComplexNumber(p1 - p2, q1-q2)
    
    def __mul__(self, other):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        p2, q2 = other.real, other.imaginary
        return ComplexNumber(p1*p2-q1*q2, p1*q2+p2*q1)

    def __truediv__(self, other):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        p2, q2 = other.real, other.imaginary
        return ComplexNumber(float(p1*p2+q1*q2)/float(p2**2+q2**2), float(q1*p2-p1*q2)/float(p2**2+q2**2))
    #Task 3
    def getReal(self):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        return p1
    
    def getImag(self):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        return q1
    #Task 4
    def is_imaginary(self):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        return q1!=0
    def is_real(self):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        return q1==0
    #Task 5
    def __repr__(self):
        return f"{self.real} + i*{self.imaginary}"
    #Task 6
    def arg_and_abs(self):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        return arctan(float(q1)/p1), sqrt(p1**2+q1**2) 
    #Task 7
    def is_equal(self, other):  # assuming other is a RationalNumber
        p1, q1 = self.real, self.imaginary
        p2, q2 = other.real, other.imaginary
        return (q1==q2) & (p1==p2)
    

    
q =  ComplexNumber(1, 2)
q1 =  ComplexNumber(1, 0)
p= ComplexNumber(3, 4)
p1= ComplexNumber(1, 2)
f=2.0
print(q)
print(q+p)
print(q-p)
print(q*p)
print(q/p)
print(q.getReal())
print(q.getImag())
print(q.is_imaginary())
print(q1.is_real())
print(q.is_equal(p))
print(q.is_equal(p1))
print(q.arg_and_abs())
print(q+f)
print(f+q)
```
