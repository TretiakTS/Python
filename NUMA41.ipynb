```python
import numpy as np
from matplotlib.pyplot import *


def coef(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:,0] = y
    #print(coef)
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
    #print(coef)        
    return coef

def newton(coef, x_data, x):
    n = len(x_data) - 1 
    p = coef[n]
    #rint(p)
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p


x_new = np.arange(-1, 1, .0001)

#func 1 plot 5 nodes
x1 = np.linspace(-1,1,5)
z2 = [np.exp(-4*i**2) for i in x1]
z3 = [np.exp(-4*i**2) for i in x_new]
y2 = coef(x1, z2)[0, :]
y_new2 = newton(y2, x1 , x_new)
plot(x_new, z3,'r')
plot(x1, z2,"o")
plot(x_new, y_new2,'y')
show()


#func 1 plot 12 nodes
x1 = np.linspace(-1,1,12)
z2 = [np.exp(-4*i**2) for i in x1]
z3 = [np.exp(-4*i**2) for i in x_new]
y2 = coef(x1, z2)[0, :]
y_new2 = newton(y2, x1, x_new)
plot(x_new, z3,'r')
plot(x1, z2,"o")
plot(x_new, y_new2,'y')
show()


#func 2 plot 15 nodes
x1 = np.linspace(-1,1,15)
z2 = [1/(1 + 25*i**2) for i in x1]
z3 = [1/(1 + 25*i**2) for i in x_new]
y2 = coef(x1, z2)[0, :]
y_new2 = newton(y2, x1, x_new)
plot(x_new, z3,'r')
plot(x1, z2,"o")
plot(x_new, y_new2,'y')
show()

#func 2 plot 21 nodes
x1 = np.linspace(-1,1,21)
z2 = [1/(1 + 25*i**2) for i in x1]
z3 = [1/(1 + 25*i**2) for i in x_new]
# get the divided difference coef
y2 = coef(x1, z2)[0, :]
y_new2 = newton(y2, x1, x_new)
plot(x_new, z3,'r')
plot(x1, z2,"o")
plot(x_new, y_new2,'y')
```
